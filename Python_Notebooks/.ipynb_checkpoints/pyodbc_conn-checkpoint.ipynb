```python
import pyodbc

cnxn = pyodbc.connect(r'DRIVER={SQL Server Native Client 11.0};SERVER=ORLEBIDEVDB;DATABASE=INTEGRATION;Trusted_Connection=yes;')
cursor = cnxn.cursor()

#Sample select query
"""
cursor.execute("SELECT @@version;") 
row = cursor.fetchone() 
while row: 
    print(row[0]) 
    row = cursor.fetchone()
"""

query = (
"""
WITH unlabeled AS
(
	SELECT dtl.unitnumber, CASE WHEN dtl.isolate IS NOT NULL THEN 1 ELSE 0 END is_qrt, dtl.reportdatesk maxReportDateSK
	FROM [INTEGRATION].[dbo].[INT_FCT_BIO_InventoryDetail] dtl 
	where dtl.reportdatesk = (select max(reportdatesk) MAX_REPDATESK from [INTEGRATION].[dbo].[INT_FCT_BIO_InventoryDetail])
	and dtl.labeldate is null 
	and dtl.rsaproductcode in ('0197013','0199013','0198043')   
)
SELECT DISTINCT 
	mr.unitnumber,
	mr.is_qrt,
	mr.is_def,
	mr.not_released,
	mr.is_abs,
	mr.collectDateSK,
	mr.maxReportDateSK,
	mr.collectAge
FROM
(
	SELECT 
		dtl.unitnumber, 
		dtl.is_qrt,
		CASE WHEN dr.deferral_code IS NOT NULL THEN 1 ELSE 0 END is_def, 
		CASE WHEN rds.Released = 'N' THEN 1 ELSE 0 END not_released,
		mkt.is_abs,
		mkt.collectDateSK, 
		dtl.maxReportDateSK,
		datediff(day, convert(varchar(10), mkt.collectDateSK, 120), convert(varchar(10),  dtl.maxReportDateSK, 120)) collectAge
	FROM unlabeled dtl 
	inner join
	(
		SELECT 
			mk.unitnumber, mk.registrationid, mk.CollectionDateSK collectDateSK, 
			CASE WHEN bt.TestResult IS NULL THEN 0 ELSE 1 END is_abs
		FROM [INTEGRATION].[dbo].[INT_MKTCollectionDetails] mk
		LEFT JOIN [INTEGRATION].[dbo].[INT_FCTBloodTestResult] bt
			ON bt.RegistrationID = mk.registrationID
			-- get abs test result
			and bt.TestResultsk = 1 -- (SELECT TESTSK FROM [INTEGRATION].[dbo].[INT_DIMTestTypes] WHERE UPPER(DESCRIPTION) = 'ABS')
	) mkt 
		on mkt.unitnumber = dtl.unitnumber  
	left join [STAGE].[dbo].[STG_RSA_DonorDeferral] dr
		on dr.registration_id = mkt.registrationID
	left join [STAGE].[dbo].[STG_RSADSReview] rds
		ON rds.registration_id = mkt.registrationID
) mr
WHERE mr.collectAge >= 6 
-- IS_QUARANTINED OR IS DEFERRAL OR IS NOT RELEASED
-- and (mr.is_qrt = 1 or mr.is_def = 1 or not_released = 1 or is_abs = 1)
order by mr.unitnumber
"""
)
    
cursor.execute(query)
    
plRecRows = cursor.fetchall()
    
for row in plRecRows:
    print(row)
```
